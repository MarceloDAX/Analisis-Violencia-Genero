```python
# Limpieza y reemplazos del dataset (en mi caso está limpio desde excel pero pongo ejemplos por si los necesitase):
#---------------------------------------------------------------------------------------------------------------------
# Reemplazar ceros en la columna 'Victima_edad' por la mediana:

def reemplazar_ceros_con_mediana(df, columna):
    mediana = df[columna].median()
    df[columna] = df[columna].replace(0, mediana)
    return df
df = reemplazar_ceros_con_mediana(df, 'Victima_edad')

#--------------------------------------------------------------------------------------
# Limpiar valores nulos en la columna 'Victima_genero' eliminando las filas con nulos:

def limpiar_nulos(df, columna, metodo='eliminar', valor_imputacion=None):
    if metodo == 'eliminar':
        df = df[df[columna].notna()]
    elif metodo == 'imputar' and valor_imputacion is not None:
        df[columna] = df[columna].fillna(valor_imputacion)
    return df
df = limpiar_nulos(df, 'Victima_genero', metodo='eliminar')

#--------------------------------------------------------------------------------------
# Unificar categorías en la columna 'Victima_genero'

def unificar_categorias(df, columna, categorias_equivalentes):
    for categoria_estandar, variantes in categorias_equivalentes.items():
        df[columna] = df[columna].replace(variantes, categoria_estandar)
    return df
categorias_dict = {
    'Femenino': ['femenino', 'FEMENINO', 'fem'],
    'Masculino': ['masculino', 'MASCULINO']
}
df = unificar_categorias(df, 'Victima_genero', categorias_dict)
```

## **04.A1 - Análisis Univariado gráfico A:**

Se  utiliza un gráfico de barras horizontales analizando la variable del dataset Violencia_tipo

```python
def graficar_variable_univariada(df, columna, titulo='Tipos de Violencia denunciadas (2017–2024)', color_palette='Set2'):
    import matplotlib.pyplot as plt
    import seaborn as sns

    conteo = df[columna].value_counts().sort_values(ascending=True)
    total = conteo.sum()
    porcentajes = (conteo / total * 100).round(2)

    fig, ax = plt.subplots(figsize=(10,6))
    bars = sns.barplot(x=conteo.values, y=conteo.index, palette=color_palette, ax=ax)

#----------------------------------------------------------------------------------------------------------------------------------
    for i, (valor, porcentaje) in enumerate(zip(conteo.values, porcentajes)):
        texto = f'{valor} ({porcentaje:.2f}%)'
        if valor > total * 0.25:  # Si la barra es "larga", escribir dentro con letra blanca
            ax.text(valor * 0.98, i, texto, va='center', ha='right', color='white', fontsize=10, fontweight='bold')
        else:
            ax.text(valor + total * 0.01, i, texto, va='center', fontsize=10)

    ax.set_title(titulo, fontsize=18, fontweight='bold', pad=15)
    ax.set_xlabel('Cantidad de denuncias')
    ax.set_ylabel(columna.replace('Violencia_tipo', 'Tipo de Violencia ').title())

#----------------------------------------------------------------------------------------------------------------------------------

    plt.figtext(0.01, -0.06, f'Total de denuncias: {total}', fontsize=14, fontweight='bold') # Agregando total de casos

    plt.tight_layout(rect=[0, 0.05, 1, 1])  #
    plt.show()
```

```python
#  Realizando el gráfico:
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

graficar_variable_univariada(df, 'Violencia_tipo')
```

## **04.A2 - Análisis Univariado gráfico B:**
Se  utiliza un gráfico de línea suavizado analizando la variable del dataset Violencia_tipo


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

#---------------------------------------------------------------------------------------------------------------------------------------

violencia_counts = df['Violencia_tipo'].value_counts()
violencia_types = violencia_counts.index
total_cases = violencia_counts.sum()

#---------------------------------------------------------------------------------------------------------------------------------------
# Coordenadas originales:

x_vals = np.arange(len(violencia_types))
y_vals = violencia_counts.values

#---------------------------------------------------------------------------------------------------------------------------------------
# Interpolación suavizada:

x_smooth = np.linspace(x_vals.min(), x_vals.max(), 300)
spl = make_interp_spline(x_vals, y_vals, k=3)
y_smooth = spl(x_smooth)

#---------------------------------------------------------------------------------------------------------------------------------------
# Creando gráfico:

plt.figure(figsize=(10, 6))

#---------------------------------------------------------------------------------------------------------------------------------------
# Línea suavizada:

plt.plot(x_smooth, y_smooth, color='darkblue', linewidth=2)

#---------------------------------------------------------------------------------------------------------------------------------------
# Puntos originales:

plt.scatter(x_vals, y_vals, color='darkblue')

#---------------------------------------------------------------------------------------------------------------------------------------
# Etiquetas en los puntos reales:

for i, val in enumerate(y_vals):
    porcentaje = (val / total_cases) * 100
    porcentaje_str = "<0.1%" if porcentaje < 0.1 else f"{porcentaje:.2f}%"
    etiqueta = f'{val:,} ({porcentaje_str})'
    plt.text(x_vals[i], y_vals[i] + 800, etiqueta,  # ajustá el +800 si necesitás más altura
             ha='center', va='bottom', fontsize=9,
             bbox=dict(facecolor='white', edgecolor='grey', boxstyle='round,pad=0.2', alpha=0.9))

#---------------------------------------------------------------------------------------------------------------------------------------
# Títulos y ajustes:

plt.title('Tipos de Violencia Denunciadas - Curva Suavizada - (2017-204)', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Tipo de Violencia', fontsize=12)
plt.ylabel('Cantidad de Denuncias', fontsize=12)
plt.xticks(x_vals, violencia_types, rotation=45, ha='right', fontsize=10)
plt.ylim(-5000, y_vals.max() * 1.25)
plt.grid(True, linestyle='--', alpha=0.8)

#---------------------------------------------------------------------------------------------------------------------------------------
# Total arriba a la derecha:

plt.text(0.98, 0.98, f'Total de denuncias: {total_cases:.0f}', ha='right', va='top', fontsize=10,
         color='darkblue', transform=plt.gca().transAxes, fontweight='bold')

#---------------------------------------------------------------------------------------------------------------------------------------
# Realizando el gráfico:

plt.tight_layout()
plt.show()
```

# **04.A3 - Análisis Univariado gráfico C:**

Se utiliza un gráfico doble (los anteriores agregados) analizando la variable del dataset Violencia_tipo

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline

#---------------------------------------------------------------------------------------------------------------------------
# Preparar datos:

violencia_counts = df['Violencia_tipo'].value_counts()
violencia_types = violencia_counts.index
total_cases = violencia_counts.sum()

#---------------------------------------------------------------------------------------------------------------------------
# === FIGURA CON DOS GRÁFICOS ===

fig, ax = plt.subplots(1, 2, figsize=(20, 8))

#---------------------------------------------------------------------------------------------------------------------------
# === GRÁFICO 1: Barras horizontales ===

conteo = violencia_counts.sort_values(ascending=True)
porcentajes = (conteo / total_cases * 100)

bars = sns.barplot(x=conteo.values, y=conteo.index, palette='Set2', ax=ax[0])

#---------------------------------------------------------------------------------------------------------------------------
# Añadiendo texto a las barras:

for i, (valor, porcentaje) in enumerate(zip(conteo.values, porcentajes.values)):
    porcentaje_str = "<0.1%" if porcentaje < 0.1 else f"{porcentaje:.2f}%"
    texto = f'{valor:,} ({porcentaje_str})'
    if valor == conteo.values.max():
        ax[0].text(valor - 2000, i, texto, va='center', ha='right', fontsize=9, color='white', fontweight='bold')
    else:
        ax[0].text(valor + 100, i, texto, va='center', ha='left', fontsize=9)

ax[0].set_title('Tipos de Violencia Denunciadas (2017–2024)', fontsize=14, fontweight='bold')
ax[0].set_xlabel('Cantidad de Casos')
ax[0].set_ylabel('')
ax[0].text(0.98, 0.95, f'Total de denuncias: {total_cases}', transform=ax[0].transAxes,
           fontsize=10, color='black', ha='right', va='top', fontweight='bold')

#---------------------------------------------------------------------------------------------------------------------------
# === GRÁFICO 2: Línea suavizada ===

x_vals = np.arange(len(violencia_types))
y_vals = violencia_counts.values
x_smooth = np.linspace(x_vals.min(), x_vals.max(), 300)
spl = make_interp_spline(x_vals, y_vals, k=3)
y_smooth = spl(x_smooth)

ax[1].plot(x_smooth, y_smooth, color='darkblue', linewidth=2)
ax[1].scatter(x_vals, y_vals, color='darkblue')

for i, val in enumerate(y_vals):
    porcentaje = (val / total_cases) * 100
    porcentaje_str = "<0.1%" if porcentaje < 0.1 else f"{porcentaje:.2f}%"
    etiqueta = f'{val:,} ({porcentaje_str})'
    ax[1].text(x_vals[i], y_vals[i] + 800, etiqueta, ha='center', fontsize=9,
               bbox=dict(facecolor='white', edgecolor='grey', boxstyle='round,pad=0.2', alpha=0.9))

ax[1].set_title('Tipos de Violencia Denunciadas - Curva Suavizada - (2017-204)', fontsize=14, fontweight='bold', color='darkblue')
ax[1].set_xlabel('Tipo de Violencia')
ax[1].set_ylabel('Cantidad de Denuncias')
ax[1].set_xticks(x_vals)
ax[1].set_xticklabels(violencia_types, rotation=45, ha='right')
ax[1].set_ylim(-5000, y_vals.max() * 1.25)
ax[1].grid(True, linestyle='--', alpha=0.8)
ax[1].text(0.98, 0.98, f'Total de denuncias: {total_cases:.0f}', ha='right', va='top', fontsize=10,
         color='darkblue', transform=plt.gca().transAxes,fontweight='bold')


#---------------------------------------------------------------------------------------------------------------------------
# Realizando los gráficos:

plt.suptitle('TIPOS DE VIOLENCIAS DENUCIADAS (2017-2024)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

# **04.a: Definicición de la variable Riesgo Alto:**

```python
## FUNCIÓN PARA CLASIFICAR RIESGO ALTO O BAJO: ##
#-------------------------------------------------------------------------------------------------------------------------------

def clasificar_riesgo(row):
    condiciones = 0
    if row['Violencia_tipo'] in ['Física', 'Sexual']:
        condiciones += 1
    if isinstance(row['Agresor_relacion'], str) and 'pareja' in row['Agresor_relacion'].lower():
        condiciones += 1
    if row['Agresor_cantidad'] >= 1:
        condiciones += 1
    if pd.notna(row['Victima_edad']) and row['Victima_edad'] <= 30:
        condiciones += 1
    if isinstance(row['Llamado_derivacion'], str) and 'sin' in row['Llamado_derivacion'].lower():
        condiciones += 1
    return 1 if condiciones >= 3 else 0

# Aplicar la función al dataset:
#-------------------------------------------------------------------------------------------------------------------------------
df['Riesgo_Alto'] = df.apply(clasificar_riesgo, axis=1)
```

#**05.e Curvas ROC y valores AUC:**

La curva ROC (Receiver Operating Characteristic) permite visualizar el desempeño de un modelo en términos de la relación entre tasa de verdaderos positivos y tasa de falsos positivos.  
El **AUC (Área Bajo la Curva)** representa qué tan bien el modelo distingue entre clases. Cuanto más cerca de 1, mejor.

A continuación se visualizan las curvas ROC de los tres modelos principales en paralelo.  
Esto permite comparar de manera directa qué tan bien separa cada modelo las clases, en términos de su capacidad predictiva (AUC).


```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

modelos_prob = {
    "Árbol de Decisión": modelo_arbol.predict_proba(X_test)[:, 1],
    "Random Forest": modelo_rf.predict_proba(X_test)[:, 1],
    "Regresión Logística": modelo_log.predict_proba(X_test)[:, 1]
}

for i, (nombre, y_scores) in enumerate(modelos_prob.items()):
    fpr, tpr, _ = roc_curve(y_test, y_scores)
    auc = roc_auc_score(y_test, y_scores)

    axes[i].plot(fpr, tpr, label=f"AUC = {auc:.4f}")
    axes[i].plot([0, 1], [0, 1], linestyle="--", color="gray")
    axes[i].set_title(nombre)
    axes[i].set_xlabel("Falsos Positivos")
    axes[i].set_ylabel("Verdaderos Positivos")
    axes[i].legend(loc="lower right")
    axes[i].grid(True)

plt.suptitle("Curvas ROC - Comparación de Modelos", fontsize=16)
plt.tight_layout()
plt.show()
```

Comparación de curvas ROC y AUC:

El siguiente gráfico muestra las curvas ROC (Receiver Operating Characteristic) de los tres modelos evaluados:

-  Árbol de Decisión  
-  Random Forest  
-  Regresión Logística  

Cada curva representa la capacidad del modelo para diferenciar entre clases (riesgo alto vs. riesgo bajo), comparando la tasa de verdaderos positivos frente a la de falsos positivos para distintos umbrales.

- Cuanto más se acerque la curva al vértice superior izquierdo, **mejor es la discriminación del modelo**.
- La métrica **AUC (Area Under the Curve)** cuantifica esa capacidad:  
  - AUC = 1.0 → modelo perfecto  
  - AUC = 0.5 → sin capacidad predictiva (equivalente a azar)

En este caso, los tres modelos presentan curvas cercanas al ideal, siendo el **Random Forest** el que logra la mayor área bajo la curva (AUC), lo que refuerza su elección como modelo final recomendado.
